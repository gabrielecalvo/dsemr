# Data
We start by loading the Mauna Loa data into Python.

The most populare library to do data analysis in python is called *pandas*. It can be installed using the command: ```pip install pandas``` from the command prompt and it can be used as follows.

````python
import pandas as pd

df = pd.read_csv("data/week2/maunaloa.csv")

df.head()
````

At this stage we are interested in the column `average` in relation to the time variables `year` and `month`. We can create a Series with TimeIndex as follows.

````python
df.index = df.apply(lambda x: pd.Timestamp(int(x.year), int(x.month), 1), axis=1)
ts = df['average']

ts.head()
````

We will now save this as an `.pickle` files, so we can go straight to the
`ts` object when we analyse the time series later on with `ts = pd.read_pickle('data/week2/maunaloa.pickle')`

````python
ts.to_pickle('data/week2/maunaloa.pickle')
````

Now we can plot the data

````python
ts.plot()
````

On FutureLearn, a few different extra configurations were applied which we can replicate in python using the following.

````python
ax = ts.plot(c='#FE40ED', style='s', ms=0.5, title="Mauna Loa CO$_2%$ levels")
ax.set_xlabel("Year")
ax.set_ylabel("Carbon dioxide level (ppmnv)");
````

Too better see the periodic nature of the time series we zoom into the period from 2010 to 2018. We do this by specifying `xlim=('2010','2018')`.

````python
ax = ts.plot(c='#FE40ED', style='s', ms=0.5, title="Mauna Loa CO$_2%$ levels", xlim=('2010','2018'))
ax.set_xlabel("Year")
ax.set_ylabel("Carbon dioxide level (ppmnv)");
````

Alternatively we can subset the time series object using the `iloc` method.

````python
ax = ts.iloc[612:718].plot(c='#FE40ED', style='s', ms=0.5, title="Mauna Loa CO$_2%$ levels")
ax.set_xlabel("Year")
ax.set_ylabel("Carbon dioxide level (ppmnv)");
````
