```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics import tsaplots
%matplotlib inline 
```

```python
plt.rcParams['figure.figsize'] = (15, 6)  # overwrite the default figure size
plt.rcParams['axes.grid'] = True          # add the grid lines by default 
```

# Purely random data
We first generate the data as Gaussian white noise.

```python
np.random.seed(0)
ser = pd.Series(np.random.normal(loc=0, scale=1, size=100))
```

```python
ser.plot(title='Time Series')
```

 and the autocorrelation function.

```python
tsaplots.plot_acf(ser, lags=25, title="Autocorrelation function");
```

# Data with a trend
We first generate the data and plot it.

```python
np.random.seed(0)
time = np.arange(1, 100+1)
ser = pd.Series(-0.1*time + np.random.normal(loc=0, scale=1, size=100))

ser.plot(title='Time Series')
```

 and the autocorrelation function.

```python
tsaplots.plot_acf(ser, lags=25, title="Autocorrelation function");
```

The linear trend leads to a very slowly decaying autocorrelation function.

# Data with a seasonal effect
We first generate data with a seasonal, in our case cosine, component and plot it

```python
np.random.seed(0)
time = np.arange(1, 100+1)
seasonal = 2*np.cos(time/(0.5*np.pi/12*10))
ser = pd.Series(seasonal + np.random.normal(loc=0, scale=1, size=100))

ser.plot(title='Time Series')
```

 and the autocorrelation function.

```python
tsaplots.plot_acf(ser, lags=25, title="Autocorrelation function");
```

# Data with a trend and a seasonal effect
We finally add both a trend and seasonal component to the simulated time series.

```python
np.random.seed(0)
time = np.arange(1, 100+1)
seasonal = 2*np.cos(time/(0.5*np.pi/12*10))
ser = pd.Series(seasonal + np.random.normal(loc=0, scale=1, size=100) + 0.1*time)

ser.plot(title='Time Series')
```

 and the autocorrelation function.

```python
tsaplots.plot_acf(ser, lags=25, title="Autocorrelation function");
```
