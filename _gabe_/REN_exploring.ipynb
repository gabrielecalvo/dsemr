```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import REN_util as util

DATA_FLD = Path('../data/week6')
[print(i) for i in DATA_FLD.glob('*.csv')];
```

# Wind Data Analysis

```python
df = pd.read_csv(DATA_FLD / 'WindData.csv', index_col='TARGETdtm', parse_dates=True)
df['ws10'], df['wd10'] = util.cart2pol(df['U10'], df['V10'])
df['ws100'], df['wd100'] = util.cart2pol(df['U100'], df['V100'])
```

```python
df.corr().style.background_gradient()
```

```python
fig, ax = plt.subplots()
df.plot.scatter('ws100', 'Power', c='wd100', s=2, alpha=0.4, cmap='jet', figsize=(20,8), grid=True, ax=ax)
```

# Solar Data Analysis

```python
df = pd.read_csv(DATA_FLD / 'SolarData.csv', index_col='TARGETdtm', parse_dates=True)
df['ws10'], df['wd10'] = util.cart2pol(df['U10'], df['V10'])
df.corr().style.background_gradient()
```

```python
fig, ax = plt.subplots()
df.plot.scatter('SurfaceRad', 'Power', c='T2', s=2, alpha=0.4, cmap='jet', figsize=(20,8), grid=True, ax=ax)
```
